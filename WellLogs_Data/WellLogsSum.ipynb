```python
import pandas as pd
import os
```

```python
# Finding excel files
all_files_in_folder = os.listdir('.')
excel_files = [file for file in all_files_in_folder if file.endswith('.xlsx')]
```

```python
summary_list = []

print(f"Found {len(excel_files)} Excel files to process...")
```

```python
for file in excel_files:

    df = pd.read_excel(file)

    well_name = df['well'].iloc[0]

    avg_density = df['Density'].mean()
    avg_porosity = df['Porosity'].mean()
    avg_perm_microd = df['perm_microd'].mean()

    file_summary = {
        'well': well_name,
        'dens': avg_density,
        'poro': avg_porosity,
        'permmicrod': avg_perm_microd
    }

    summary_list.append(file_summary)

print("Processing complete!")
```

```python
final_summary_df = pd.DataFrame(summary_list)

output_filename = 'wells_summary_output.xlsx'

final_summary_df = final_summary_df.sort_values(by='well', ignore_index=True)

final_summary_df.to_excel(output_filename, index=False)

print(f"\n Summary table successfully saved as '{output_filename}'")
```
